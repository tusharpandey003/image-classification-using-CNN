```python
import os
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import load_model
```

```python
yhat=load_model(os.path.join('models','image_detection.h5'))
```

```python
img=cv2.imread('test1.jpeg')
plt.imshow(img)
plt.show()
```

```python
resize=tf.image.resize(img,(256,256))
plt.imshow(resize.numpy().astype(int))
```

```python
yhatnew=yhat.predict(np.expand_dims(resize/255,0))
yhatnew
```

```python
if yhatnew > 0.5: 
    print(f'Predicted class is normal')
else:
    print(f'Predicted class is affected')
```
